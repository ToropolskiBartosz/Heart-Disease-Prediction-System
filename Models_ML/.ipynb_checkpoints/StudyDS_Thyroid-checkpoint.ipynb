```python
from moduls.load_data import DataMenager
import os
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
import seaborn as sns
```

# Ładowanie danych

```python
DOWNLOAD_ROOT = 'https://archive.ics.uci.edu/ml/machine-learning-databases'
DATA_LOCATION_TRAIN = '/thyroid-disease/ann-train.data'
DATA_LOCATION_TEST = '/thyroid-disease/ann-test.data'
DATA_URL_TRAIN = DOWNLOAD_ROOT + DATA_LOCATION_TRAIN
DATA_URL_TEST = DOWNLOAD_ROOT + DATA_LOCATION_TEST
DATA_PATH = 'data_set'
```

```python
'''
Nadanie kolumną odpowiednich nazw 
względem podanych informacji na stronie :
https://www.is.umk.pl/~duch/projects/projects/datasets.html
'''

columns = ['age', 'sex', 'on_thyroxine', 
           'maybe_on_thyroxine', 'on_antithyroid_medication', 'sick_patient_reports_malaise',
           'pregnant', 'thyroid_surgery', 'treatment',
           'test_hypothyroid', 'test_hyperthyroid', 'on_lithium',
           'has_goitre','has_tumor','hypopituitary',
           'psychological_symptoms','tsh','t_three',
           'tt_four','t_four_u','fti',
           'target']
#Definicja kolumn z wartościmi dyskretnymi i ciągłymi 
cat_col = [ 'sex', 'on_thyroxine', 'maybe_on_thyroxine', 
           'on_antithyroid_medication', 'sick_patient_reports_malaise','pregnant', 
           'thyroid_surgery', 'treatment','test_hypothyroid', 'test_hyperthyroid', 
           'on_lithium','has_goitre','has_tumor','hypopituitary',
           'psychological_symptoms','target']

num_col = ['tsh','t_three', 'tt_four','t_four_u','fti','age']
```

```python
dm_train = DataMenager(DATA_PATH, DATA_URL_TRAIN)# Tworzenie istancji własnej klasy
dm_test = DataMenager(DATA_PATH, DATA_URL_TEST)# Tworzenie istancji własnej klasy
```

```python
dm_train.fetch_data_from_web('thyroid_train.csv',' ')
dm_test.fetch_data_from_web('thyroid_test.csv',' ')
```

```python
raw_df_train_web = dm_train.load_raw_data('thyroid_train.csv')
raw_df_test_web = dm_train.load_raw_data('thyroid_test.csv')
```

```python
raw_df_train = raw_df_train_web.iloc[:,:-2]
raw_df_train.columns = columns
raw_df_train.head()
```

```python
raw_df_test = raw_df_test_web.iloc[:,:-2]
raw_df_test.columns = columns
raw_df_test.head()
```

```python
raw_df = pd.concat([raw_df_train,raw_df_test])
raw_df['age'] = raw_df['age']*100
```

```python
df = raw_df.copy()
```

# Usuwanie

```python
#Sprawdzanie czy istnieją braki w danch
df.isnull().sum()
```

```python
#Sprawdzenie szczegółowe zawartości kolumn typu object
for i in cat_col:
    print(50*'-'+'\n' + str(df[f'{i}'].value_counts()));
```

```python
plt.figure(figsize=(10,8))
corr_df = df[num_col].corr()
sns.heatmap(corr_df,annot=True)
```

```python
#Rozrzut danych względem targetu
plt.figure(figsize=(20,20))
for i,feature in enumerate(num_col):
    plt.subplot(4,2,i+1)
    sns.boxplot(x=feature,  data=df, linewidth=2.5)
```

```python
#Usunięcie rekordów z pustymi komórkami
df = df.dropna()
```

```python
transformationDict = {3: 0, 1: 1, 2: 1}
df['target'] = df['target'].map(transformationDict)
```

# Zmiana typów danych kolumn

```python
for col in cat_col:
    df[col] = df[col].astype('category')
df.info()
```

# Opis danych 

## Opis danych dyskretnych

```python
#Opis danych dyskretnych
df.describe(include=['category']).T
```

```python
#Wykresy 
plt.figure(figsize=(20,40))
for i,feature in enumerate(cat_col):
    plt.subplot(8,3,i+1)
    ax = sns.countplot(x=feature,data=df)
    for p in ax.patches:
        ax.annotate(f'\n{p.get_height()+20}', 
                        (p.get_x()+0.2, p.get_height()+5),
                        ha='center', va='top', color='black', size=11)
```

```python
#Wykres kołowy służaący do reprezentacji stosunku
plt.figure(figsize=(20,30))
for i,feature in enumerate(cat_col):
    plt.subplot(7,3,i+1)
    df[feature].value_counts().plot(kind='pie', autopct="%.1f%%")
```

```python
#Wykres reprezentujący wartości kolumn dyskretnych względem targetu
plt.figure(figsize=(20,40))
for i,feature in enumerate(cat_col):
    if feature != 'target':
        plt.subplot(7,3,i+1)
        ax = sns.countplot(x=feature,hue='target',data=df)
        for p in ax.patches:
            ax.annotate(f'\n{p.get_height()+20}', 
                        (p.get_x()+0.2, p.get_height()+5),
                        ha='center', va='top', color='black', size=11)
```

## Opis danych ciągłych

```python
#Opis danych
df.describe().T
```

```python
#Wykres reprezentujący rozkład danych ciągłych
g = sns.pairplot(df, diag_kind="hist", height=2.5)
g.map_lower(sns.kdeplot, levels=4, color=".2")
```

```python
#Wykres reprezentujący rozkład danych ciągłych względem targetu
t = sns.pairplot(df, hue='target', diag_kind="hist", height=2.5)
```

```python
#Rozrzut danych względem targetu
plt.figure(figsize=(20,20))
for i,feature in enumerate(num_col):
    plt.subplot(4,2,i+1)
    sns.boxplot(x=feature, y='target',  data=df, linewidth=2.5)
```
