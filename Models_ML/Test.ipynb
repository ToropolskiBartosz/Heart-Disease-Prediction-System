```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,roc_auc_score,classification_report,f1_score
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler 
```

```python
df = pd.read_csv('./dane/HeartDiseaseDataSet.csv') 
df.head()
```

```python
df_dummies = pd.get_dummies(data=df,drop_first=True)
df_dummies.head()
```

```python
df_dummies.shape
```

```python
num_cols=['age','trestbps','chol','thalach','oldpeak','ca']
```

```python
y = df_dummies["target"]
X = df_dummies.drop('target',axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state = 0)

scaler = StandardScaler()
X_train.loc[:,num_cols] = scaler.fit_transform(X_train.loc[:,num_cols])

X_test.loc[:,num_cols] = scaler.transform(X_test.loc[:, num_cols])
```

```python
X_train
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(X_train, y_train)
y_pred_lr = classifier.predict(X_test)

lr_acc_score=metrics.accuracy_score(y_test, y_pred_lr)
print('Test Accuracy score: ', lr_acc_score)
lr_conf_matrix = confusion_matrix(y_test, y_pred_lr)
print("confusion matrix")
print(lr_conf_matrix)
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier()
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred_dt = classifier.predict(X_test)

dt_acc_score=metrics.accuracy_score(y_test, y_pred_dt)
print('Test Accuracy score: ', dt_acc_score)
dt_conf_matrix = confusion_matrix(y_test, y_pred_dt)
print("confusion matrix")
print(dt_conf_matrix)
```

```python
knn = KNeighborsClassifier(metric='manhattan', n_neighbors=3)
knn.fit(X_train, y_train)
y_pred_knn = knn.predict(X_test)
knn_acc_score=metrics.accuracy_score(y_test, y_pred_knn)
print(f'{knn.score(X_test,y_test)}')
print('Test Accuracy score: ', knn_acc_score)
knn_conf_matrix = confusion_matrix(y_test, y_pred_knn)
print("confusion matrix")
print(knn_conf_matrix)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10)
classifier.fit(X_train, y_train)

# Predicting the Test set results
y_pred = classifier.predict(X_test)

acc_score=metrics.accuracy_score(y_test, y_pred)
print('Test Accuracy score: ', acc_score)
conf_matrix = confusion_matrix(y_test, y_pred_knn)
print("confusion matrix")
print(conf_matrix)
```
